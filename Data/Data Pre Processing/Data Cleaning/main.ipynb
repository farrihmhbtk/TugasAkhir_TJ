```python
# version 1.2

import csv
import re
import string
import html

def clean_tweet(text):
    """
    Clean a tweet by applying the following transformations:
    - Remove URLs (starting with http, https, www).
    - Remove all emoji.
    - Unescape HTML entities (e.g., '&amp;' becomes '&').
    - Remove all non-ASCII characters (to eliminate any weird characters).
    - Process mentions: keep only the mention for 'pt_transjakarta' (standardized to lower-case) and remove all others.
    - Remove the '#' symbol from hashtags (keeping the hashtag text).
    - Separate every punctuation mark from adjacent characters with whitespace.
    - Separate the Indonesian prefix 'di' from the main word (case-insensitive).
    - Separate the Indonesian suffix 'nya' from the main word (case-insensitive).
    - Separate letters and numbers if they are attached (e.g., 'k13' becomes 'k 13'), while leaving pure numbers unchanged.
    - Replace the word 'jaklingko' with 'angkutan'.
    - Remove extra whitespace.
    - Leave the original casing (do not lowercase the entire tweet, except for ensuring pt_transjakarta is standardized).
    """

    # --- Remove URLs ---
    text = re.sub(r'http\S+|www\.\S+', '', text)
    
    # --- Remove Emoji ---
    emoji_pattern = re.compile(
        "[" 
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)
    
    # --- Unescape HTML entities ---
    text = html.unescape(text)
    
    # --- Remove all non-ASCII characters (i.e. weird characters) ---
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    
    # --- Process mentions: Remove all mentions except for '@pt_transjakarta'
    def replace_mention(match):
        mention = match.group(1)
        if mention.lower() == "pt_transjakarta":
            return "pt_transjakarta"  # standardize to lower-case for pt_transjakarta
        else:
            return ""  # remove other mentioned accounts
    text = re.sub(r'@(\w+)', replace_mention, text)
    
    # --- Process hashtags: Remove '#' symbol (keeping the hashtag text)
    text = re.sub(r'#', '', text)
    
    # --- Separate punctuation from adjoining characters ---
    text = re.sub(r'([{}])'.format(re.escape(string.punctuation)), r' \1 ', text)
    text = re.sub(r'\s+', ' ', text).strip()  # normalize whitespace
    
    # --- Separate the prefix "di" from the main word (handles both 'di' and 'Di')
    text = re.sub(r'\b([dD]i)([A-Za-z]+)', r'\1 \2', text)
    
    # --- Separate the suffix "nya" from the main word (case-insensitive) ---
    text = re.sub(r'([A-Za-z]+)(nya)\b', r'\1 \2', text, flags=re.IGNORECASE)
    
    # --- Separate letters and numbers attached ---
    text = re.sub(r'([A-Za-z])(\d+)', r'\1 \2', text)
    text = re.sub(r'(\d+)([A-Za-z])', r'\1 \2', text)
    
    # --- Replace the word 'jaklingko' with 'angkutan' ---
    text = re.sub(r'\bjaklingko\b', 'angkutan', text, flags=re.IGNORECASE)
    
    # --- Final cleanup: Remove any extra whitespace ---
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Define input and output file paths
input_file = "ready v1.0/PRDECT-ID Dataset v1.0_RAW.csv"
output_file = "ready v1.0/PRDECT-ID Dataset v1.0_READY TO BE LABELED.csv"

# Open the input CSV file and read its content
with open(input_file, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    # Prepare fieldnames by adding new columns "original_tweet" and "cleaned_full_text"
    fieldnames = reader.fieldnames + ["original_tweet", "cleaned_full_text"]
    
    # Create a list to hold the updated rows
    updated_rows = []
    for row in reader:
        # Save the original tweet text (assuming the column is named "full_text")
        row["original_tweet"] = row["full_text"]
        # Clean the tweet and store the result in the new column
        row["cleaned_full_text"] = clean_tweet(row["full_text"])
        updated_rows.append(row)

# Write the updated rows to the output CSV file
with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(updated_rows)

print("File processed and saved to", output_file)
```

```python
# version 1.3

import csv
import re
import string
import html

def clean_tweet(text):
    """
    Clean a tweet by applying the following transformations:
    - Remove URLs (starting with http, https, www).
    - Remove all emoji.
    - Unescape HTML entities (e.g., '&amp;' becomes '&').
    - Remove all non-ASCII characters (to eliminate any weird characters).
    - Process mentions: keep only the mention for 'pt_transjakarta' (standardized to lower-case) and remove all others.
    - Remove the '#' symbol from hashtags (keeping the hashtag text).
    - Separate every punctuation mark (except underscore) from adjacent characters with whitespace.
    - Separate the Indonesian prefix 'di' from the main word (case-insensitive).
    - Separate the Indonesian suffix 'nya' from the main word (case-insensitive).
    - Separate letters and numbers if they are attached (e.g., 'k13' becomes 'k 13'), while leaving pure numbers unchanged.
    - Replace the word 'jaklingko' with 'angkutan'.
    - Remove extra whitespace.
    - Leave the original casing (do not lowercase the entire tweet, except for ensuring pt_transjakarta is standardized).
    """
    
    # --- Remove URLs ---
    text = re.sub(r'http\S+|www\.\S+', '', text)
    
    # --- Remove Emoji ---
    emoji_pattern = re.compile(
        "[" 
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)
    
    # --- Unescape HTML entities ---
    text = html.unescape(text)
    
    # --- Remove all non-ASCII characters (i.e. weird characters) ---
    text = re.sub(r'[^\x00-\x7F]+', '', text)
    
    # --- Process hashtags: Remove '#' symbol (keeping the hashtag text)
    text = re.sub(r'#\w+', '', text)
    
    # --- Process mentions: Keep @pt_transjakarta as 'pt_transjakarta', remove all other @mentions ---
    def process_mentions(match):
        mention = match.group(0)
        if mention.lower() == '@pt_transjakarta':
            return 'pt_transjakarta'
        else:
            return ''
    
    text = re.sub(r'@\w+', process_mentions, text, flags=re.IGNORECASE)
    
    # --- Separate punctuation from adjoining characters (exclude underscore) ---
    # Remove underscore from punctuation set
    punct_no_underscore = string.punctuation.replace('_', '')
    text = re.sub(r'([{}])'.format(re.escape(punct_no_underscore)), r' \1 ', text)
    text = re.sub(r'\s+', ' ', text).strip()  # normalize whitespace
    
    # --- Final cleanup: Remove any extra whitespace ---
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text

# Define input and output file paths
# input_file = "../../Production/twitterapi IO/self_training/batch 7 8 9.csv"
# output_file = "../../Production/twitterapi IO/self_training/self_training_batch 7 8 9-Ready to be Labeled.csv"

input_file = "../../Production/twitterapi IO/Proof of Concept/POC_Merged_deduplicated_keyword filtered.csv"
output_file = "../../Production/twitterapi IO/Proof of Concept/POC_Merged_deduplicated_keyword filtered_cleaned.csv"

# Open the input CSV file and read its content
with open(input_file, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    # Prepare fieldnames by adding new columns "original_tweet" and "cleaned_full_text"
    fieldnames = reader.fieldnames + ["original_tweet", "cleaned_full_text"]
    
    # Create a list to hold the updated rows
    updated_rows = []
    for row in reader:
        # Save the original tweet text (assuming the column is named "full_text")
        row["original_tweet"] = row["full_text"]
        # Clean the tweet and store the result in the new column
        row["cleaned_full_text"] = clean_tweet(row["full_text"])
        updated_rows.append(row)

# Write the updated rows to the output CSV file
with open(output_file, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(updated_rows)

print("File processed and saved to", output_file)
```
